```python
engine = create_engine('postgresql://username: password@localhost:port number/'test table') 
# replace username, password with your credentials and port number too. create a new database in your preferred choice(postgre,mysql,etc) and specify it in place of test table here.
```

```python
age22.to_sql('test_table',engine) # writing to sql, new table will be created , we get error if table already exists
```

```python
age22.to_sql('test_table',engine, if_exists='replace') # new table will be created , we won't get any error if table already exists, original table will be replaced
```

```python
sql_df = pd.read_sql('test_table',engine,index_col='PassengerId') # reading sql
```

```python
sql_df.head()
```

```python
sql_df = pd.read_sql_query('Select * from test_table',engine,index_col='PassengerId') # to query on this dataframe
```
