```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
pivot1 = pd.pivot_table(df, values='Fare', index='Pclass')
```

```python
pivot1
```

```python
pivot2 = pd.pivot_table(df, values='Fare', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
pivot2
```

```python
pivot2.reset_index()
```

```python
pivot3 = pd.pivot_table(df, values='Survived', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
pivot3
```

```python
pivot4 = pd.pivot_table(df, values='PassengerId', index='Pclass', columns='Embarked', aggfunc='count')
```

```python
pivot4
```

```python
pivot5 = pd.pivot_table(df, values='PassengerId', index='Pclass', aggfunc=['count','mean','sum'])
```

```python
pivot5
```

```python
pivot6 = pd.pivot_table(df, values='Fare', index='Pclass', aggfunc=['count','mean','sum'],fill_value=0)
```

```python
pivot6
```

```python
pivot7 = pd.pivot_table(df, values='Fare', index=['Pclass','Cabin'], aggfunc=['count','mean','sum'],fill_value=0)
```

```python
pivot7
```

```python
pivot8 = pd.pivot_table(df, values='Fare', index='Pclass', aggfunc='sum', margins=True, margins_name='Total')
```

```python
pivot8
```

```python
pivot9 = pd.pivot_table(df, values='Fare', index='Cabin',columns='Sex', aggfunc= lambda x:len(x))
```

```python
pivot9
```

```python
pivot10 = pd.pivot_table(df, values='Fare', index='Cabin',columns='Sex', aggfunc= lambda x:len(x),fill_value=0)
```

```python
pivot10
```

```python
pivot11 = pd.pivot_table(df, values=['Survived','Fare'], index='Cabin',columns='Sex', aggfunc = {'Survived':'sum','Fare':'mean'},fill_value=0)
```
